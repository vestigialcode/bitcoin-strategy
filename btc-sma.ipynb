```python
import numpy as np
import pandas as pd

#Data Source
import yfinance as yf

#Data viz
import plotly.graph_objs as go
```

```python
data = yf.download(tickers='BTC-INR',period = '8d', interval = '90m')
```

```python
print(data.to_string())
```

```python
data['MA5'] = data['Close'].rolling(5).mean()
data['MA20'] = data['Close'].rolling(20).mean()
```

```python
print(data.to_string())
```

```python
data.drop(['Volume','Low'], axis = 1, inplace = True)
```

```python
print(data.to_string())
```

```python
data.dropna(inplace=True)
```

```python
print(data.to_string())
```

```python
# Test for buy-sell signal
profit=0
for i in data.index:
    if data['MA5'][i]>data['MA20'][i]:
        print("buy signal")
        profit1=data['Close'][i]-data['Open'][i]
        profit=profit1+profit
    elif data['MA5'][i]<data['MA20'][i]:
        print("Sell signal")
        profit2=data['Open'][i]-data['Close'][i]
        profit=profit+profit2
        
print("Our profit is",profit)
#print("Rate of return{profit/data['Close'][0]})
```

```python
# Rate of return 
print(profit/data['Close'][0]*100)
```
